```python
def get_aliases(metric):
    if metric == "Applied to Leased Ratio":
        alias = ["Applied to Leased Ratio", "A2L %", "A2L Ratio", "Applied to Leased %"] 
    elif metric == "Toured to Applied Ratio":
        alias = ["Toured to Applied Ratio", "Closing Ratio", "T2A %", "Toured to Applied %", "T2A Ratio"]
    elif metric == "Lead to Toured Ratio":
        alias = ["Lead to Toured Ratio", "Conversion Ratio", "L2T %", "L2T Ratio", "Lead to Toured %", 
                      "Contact to Toured %", "Contact to Toured Ratio", 
                      "Prospect to Toured Ratio", "Prospect to Toured %"]
    elif metric == "Property Management Company":
        alias = ["Property Management Company", "Org Name", "PMC", "Organization", "Org", 
                      "Property Management Companies"]
    elif metric == "Property Name":
        alias = ["Property Name", "Community"]
    elif metric == "Conversion Rate":
        alias = ["Conversion Rate", "Conversion Ratio"]
    elif metric == "Conversion Velocity":
        alias = ["Conversion Velocity", "Conversion Duration"]
    elif metric == "Date Range":
        alias = ["Date Range"]
    elif metric == "All Dimensions":
        alias = ["All Dimensions", "Other Dimensions"]
    elif metric == "Metric Types (All Durations)":
        alias = ["Metric Types (All Durations)", "Other Metrics", "All Metrics"]
    elif metric == "Metric Types (All Ratios)":
        alias = ["Metric Types (All Ratios)", "Other Metrics", "All Metrics"]
    elif metric == "Applied to Leased Duration":
        alias = ["Applied to Leased Duration", "A2L Duration"]
    elif metric == "Toured to Applied Duration":
        alias = ["Toured to Applied Duration", "Closing Duration", "T2A Duration"]
    elif metric == "Lead to Toured Duration":
        alias = ["Lead to Toured Duration", "Conversion Duration", "L2T Duration", 
                      "Contact to Toured Duration", "Prospect to Toured Duration"]
    return alias
```

```python
def get_questions(metadata_column_name, hierarchy, alias):
    if metadata_column_name =="Lede + Lede.Example":
        if hierarchy == "Property Level":
            question = ["How do you define "+alias+" ?", 
                        "What does "+alias+" represent?", 
                        "Describe what "+alias+" is", 
                        "State the meaning of "+alias, 
                        "What is "+alias]
        elif hierarchy == "PMC Level":
            question = ["How do you define the benchmark for "+alias+" ?", 
                        "What does the benchmark for "+alias+" represent for my organization?", 
                        "Describe what the benchmark for "+alias+" is", 
                        "State the meaning of the benchmark for "+alias+" for my PMC", 
                        "What is the benchmark for "+alias]
        elif hierarchy == "Multifamily Level":
            question = ["How do you define the benchmark for "+alias+" in this industry?", 
                        "What does the benchmark for "+alias+" represent in multifamily?", 
                        "Describe what the benchmark for "+alias+" in this industry is", \
                        "State the meaning of the benchmark for "+alias+" in the multifamily industry", 
                        "What is the benchmark for "+alias+" in multifamily industry?"]
    elif metadata_column_name =="Lede.Example":
        if hierarchy == "Property Level":
            question = ["Provide an example of "+alias, 
                        "How can "+alias+" be exemplified?",
                        "Illustrate "+alias, 
                        "What is a case in point for "+alias+"?",
                        "How can you demonstrate "+alias+"?"]
        elif hierarchy == "PMC Level":
            question = ["Provide an example of the benchmark for "+alias, 
                        "How can the benchmark for "+alias+" be exemplified?",
                        "Illustrate the benchmark for "+alias, 
                        "What is a case in point for the benchmark for "+alias+"?",
                        "How can you demonstrate the benchmark for "+alias+"?"]
        elif hierarchy == "Multifamily Level":
            question = ["Provide an example of the benchmark for "+alias+" in this industry", 
                        "How can the benchmark for "+alias+" be exemplified in multifamily?",
                        "Illustrate the benchmark for "+alias+" in multifamily", 
                        "What is a case in point for the benchmark for "+alias+"in our industry?",
                        "How can you demonstrate the benchmark for "+alias+" in the multifamily industry?"]
    elif metadata_column_name =="Why it matters?":
        if hierarchy == "Property Level":
            question = ["Why is "+alias+" important?", 
                        "How is "+alias+" applied from a marketing standpoint?",
                        "What is business use case of "+alias+"?", 
                        "What is the value of "+alias+"?",
                        "Why do we need to measure "+alias+"?"]
        elif hierarchy == "PMC Level":
            question = ["Why is the benchmark for "+alias+" important?", 
                        "How is the benchmark for "+alias+" applied in my PMC from a marketing standpoint?",
                        "What is business use case of the benchmark for "+alias+"?", 
                        "What is the value of the benchmark for "+alias+" in my organization?",
                        "Why do we need to measure the benchmark for "+alias+"?"]
        elif hierarchy == "Multifamily Level":
            question = ["Why is the benchmark for "+alias+" important in multifamily?", 
                        "How is the benchmark for "+alias+" applied in this industry from a marketing standpoint?",
                        "What is business use case of the benchmark for "+alias+" in the multifamily industry?", 
                        "What is the value of the benchmark for "+alias+" in multifamily?",
                        "Why do we need to measure the benchmark for "+alias+" in this industry?"]
    elif metadata_column_name =="Godeeper + Godeeper.Example":
        if hierarchy == "Property Level":
            question = ["Elaborate on "+alias, 
                        "Give me a detailed explanation of "+alias+".",
                        "How do I calculate "+alias+"?", 
                        "Go deeper about "+alias, 
                        "Provide a thorough description and computation of "+alias]
        elif hierarchy == "PMC Level":
            question = ["Elaborate on the benchmark for "+alias, 
                        "Give me a detailed explanation of the benchmark for "+alias+".",
                        "How do I calculate the benchmark for "+alias+" in my PMC?", "Go deeper about the benchmark for "+alias+" in my organization", 
                        "Provide a thorough description and computation of the benchmark for "+alias]
        elif hierarchy == "Multifamily Level":
            question = ["Elaborate on the benchmark for "+alias+" in this industry", 
                        "Give me a detailed explanation of the benchmark for "+alias+"in multifamily.",
                        "How do I calculate the benchmark for "+alias+" in this industry?", 
                        "Go deeper about the benchmark for "+alias+" in multifamily industry", 
                        "Provide a thorough description and computation of the benchmark for "+alias+"in multifamily"]
    elif metadata_column_name =="Godeeper.Example":
        if hierarchy == "Property Level":
            question = ["Show me a thorough demonstration of "+alias, 
                        "Give me a detailed example of "+alias,
                        "What is a clear and detailed illustration of "+alias+"?", 
                        "How can you illustrate "+alias+" with a concrete case?", 
                        "Give me a comprehensive example of how "+alias+" is derived"]
        elif hierarchy == "PMC Level":
            question = ["Show me a thorough demonstration of the benchmark for "+alias, 
                        "Give me a detailed example of the benchmark for "+alias,
                        "What is a clear and detailed illustration of the benchmark for "+alias+" in case of my PMC?", 
                        "How can you illustrate the benchmark for "+alias+" in my organization with a concrete case?", 
                        "Give me a comprehensive example of how the benchmark for "+alias+" is derived"]
        elif hierarchy == "Multifamily Level":
            question = ["Show me a thorough demonstration of the benchmark for "+alias+" in multifamily", 
                        "Give me a detailed example of the benchmark for "+alias+" in the multifamily industry",
                        "What is a clear and detailed illustration of the benchmark for "+alias+" in this industry?", 
                        "How can you illustrate the benchmark for "+alias+" in this industry with a concrete case?", 
                        "Give me a comprehensive example of how the benchmark for "+alias+" is derived in the multifamily industry"]
    return question
```

```python
def get_metric_name(metric, hierarchy):
    if metric == 'Applied to Leased Duration':
        if hierarchy == 'Property Level':
            return "applied-to-leased-duration-metric"
        elif hierarchy == 'PMC Level':
            return "applied-to-leased-duration-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "applied-to-leased-duration-benchmark-multifamily"
    elif metric == 'Applied to Leased Ratio':
        if hierarchy == 'Property Level':
            return "applied-to-leased-ratio-metric"
        elif hierarchy == 'PMC Level':
            return "applied-to-leased-ratio-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "applied-to-leased-ratio-benchmark-multifamily"
    elif metric == 'Lead to Toured Duration':
        if hierarchy == 'Property Level':
            return "lead-to-toured-duration-metric"
        elif hierarchy == 'PMC Level':
            return "lead-to-toured-duration-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "lead-to-toured-duration-benchmark-multifamily"
    elif metric == 'Lead to Toured Ratio':
        if hierarchy == 'Property Level':
            return "lead-to-toured-duration-benchmark-multifamily"
        elif hierarchy == 'PMC Level':
            return "lead-to-toured-ratio-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "lead-to-toured-ratio-benchmark-multifamily"
    elif metric == 'Toured to Applied Duration':
        if hierarchy == 'Property Level':
            return "toured-to-applied-duration-benchmark-multifamily"
        elif hierarchy == 'PMC Level':
            return "toured-to-applied-duration-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "toured-to-applied-duration-metric"
    elif metric == 'Toured to Applied Ratio':
        if hierarchy == 'Property Level':
            return "toured-to-applied-ratio-benchmark-multifamily"
        elif hierarchy == 'PMC Level':
            return "toured-to-applied-ratio-benchmark"
        elif hierarchy == 'Multifamily Level':
            return "toured-to-applied-ratio-metric"
```

```python
dataset['metric_name_requested'] = dataset.apply(lambda x: get_metric_name(x['metric'], x['hierarchy']), axis=1)
```

```python
def get_information_requested(metadata_column_name):
    if metadata_column_name =="Lede + Lede.Example":
        return ["simple-explanation", "simple-example"]
    elif metadata_column_name =="Lede.Example":
        return ["simple-example"]
    elif metadata_column_name =="Why it matters?":
        return ["why-it-matters"]
    elif metadata_column_name =="Godeeper + Godeeper.Example":
        return ["calculation-explanation", "calculation-example"]
    elif metadata_column_name =="Godeeper.Example":
        return ["calculation-example"]
```

```python
dataset['information_requested'] = dataset['metadata_column_name'].apply(lambda x: get_information_requested(x))
```

```python
def get_output(question, metric_name_requested, information_requested):
    return {"question": question, "metric_name_requested": metric_name_requested, "information_requested": information_requested}
```

```python
dataset['output'] = dataset.apply(lambda x: get_output(x['question'], x['metric_name_requested'], x['information_requested']), axis=1)
```
