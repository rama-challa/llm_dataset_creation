```python
def get_aliases(metric):
    if metric == "Applied to Leased Ratio":
        alias = ["Applied to Leased Ratio", 
                 "Applied to Leased %", 
                 "Applied to Leased Percent",
                 "A2L %", 
                 "A2L Percent",
                 "A2L Ratio"] 
    elif metric == "Toured to Applied Ratio":
        alias = ["Toured to Applied Ratio", 
                 "Toured to Applied %", 
                 "Toured to Applied Percent",
                 "Closing Ratio", 
                 "T2A %", 
                 "T2A Percent", 
                 "T2A Ratio"]
    elif metric == "Lead to Toured Ratio":
        alias = ["Lead to Toured Ratio", 
                 "Lead to Toured %", 
                 "Lead to Toured Percent", 
                 "Conversion Ratio", 
                 "L2T %", 
                 "L2T Percent", 
                 "L2T Ratio", 
                 "Contact to Toured %", 
                 "Contact to Toured Percent",
                 "Contact to Toured Ratio", 
                 "Prospect to Toured Ratio", 
                 "Prospect to Toured %",
                 "Prospect to Toured Percent"]
    elif metric == "Applied to Leased Duration":
        alias = ["Applied to Leased Duration", 
                 "A2L Duration"]
    elif metric == "Toured to Applied Duration":
        alias = ["Toured to Applied Duration", 
                 "Closing Duration", 
                 "T2A Duration"]
    elif metric == "Lead to Toured Duration":
        alias = ["Lead to Toured Duration", 
                 "Conversion Duration", 
                 "L2T Duration", 
                 "Contact to Toured Duration", 
                 "Prospect to Toured Duration"]
    return alias
```

```python
def get_questions(metadata_column_name, hierarchy, alias):
    if metadata_column_name =="Lede + Lede.Example":
        if hierarchy == "Property Level":
            question = ["How do you define "+alias+" ?", 
                        "What does "+alias+" represent for my property?",  
                        "Describe what "+alias+" is for my community",
                        "What is "+alias]
        elif hierarchy == "PMC Level":
            question = ["How do you define the benchmark of "+alias+" for my portfolio?", 
                        "What does the average of "+alias+" represent for my community.", 
                        "Describe what the benchmark of "+alias+" is for my organization?",
                        "Describe what the benchmark of "+alias+" is for my org?",
                        "What is the benchmark for "+alias,
                        "Simplify "+alias+" for my PMC",
                        "What is the definition of "+alias+" for my property management company?"]
        elif hierarchy == "Multifamily Level":
            question = ["How do you define the benchmark of "+alias+" for multifamily?", 
                        "What does the benchmark of "+alias+" represent for this industry.", 
                        "Describe what the benchmark of "+alias+" is for multi-family industry?",
                        "What is the multifamily benchmark for "+alias]
    elif metadata_column_name =="Lede.Example":
        if hierarchy == "Property Level":
            question = ["Provide an example of "+alias, 
                        "How can "+alias+" be exemplified for my property?",
                        "Illustrate "+alias+" for my community",
                        "How can you demonstrate "+alias+"?"]
        elif hierarchy == "PMC Level":
            question = ["Provide an example of the benchmark for "+alias+" for my portfolio", 
                        "How can the average for "+alias+" be exemplified for my community?",
                        "Illustrate the benchmark for "+alias+" for my organization",
                        "Illustrate the benchmark for "+alias+" for my org",
                        "How can you demonstrate the average for "+alias+"?",
                        "Give me an illustration of the benchmark for "+alias+" for my PMC",
                        "Show me a case in point for the benchmark of "+alias+" for my property management company"]
        elif hierarchy == "Multifamily Level":
            question = ["Provide an example of the benchmark for "+alias+" in this industry", 
                        "How can the benchmark for "+alias+" be exemplified in multifamily?",
                        "Illustrate the benchmark for "+alias+" in multifamily industry",
                        "How can you demonstrate the multifamily benchmark for "+alias+"?"]
    elif metadata_column_name =="Why it matters?":
        if hierarchy == "Property Level":
            question = ["Why is "+alias+" important for my property?", 
                        "How is "+alias+" applied from a marketing standpoint for my community?",
                        "Why do we need to measure "+alias+"?",
                        "What is the value in measuring "+alias+"?"]
        elif hierarchy == "PMC Level":
            question = ["Why is the average for "+alias+" important for my portfolio?", 
                        "How is the benchmark for "+alias+" applied in my community from a marketing standpoint?",
                        "Why do we need to measure the average for "+alias+" in my organization?",
                        "Why do we need to measure the average for "+alias+" in my org?",
                        "What is the value in measuring the benchmark for "+alias+" ?",
                        "What is the purpose of calculating the benchmark for "+alias+" for my PMC?",
                        "Why do I need to know the benchmark for "+alias+" for my property management company?"]
        elif hierarchy == "Multifamily Level":
            question = ["Why is the benchmark for "+alias+" important in multifamily?", 
                        "How is the benchmark for "+alias+" applied in this industry from a marketing standpoint?",
                        "Why do we need to measure the benchmark for "+alias+" in multifamily industry?",
                        "What is the value in measuring the multifamily benchmark for "+alias+"?"]
    elif metadata_column_name =="Godeeper + Godeeper.Example":
        if hierarchy == "Property Level":
            question = ["Elaborate on "+alias+" for my property", 
                        "Give me a detailed explanation of "+alias+" for my community.",
                        "How do I calculate "+alias+"?",
                        "Provide a thorough description and computation of "+alias]
        elif hierarchy == "PMC Level":
            question = ["Elaborate on the average for "+alias+" for my community", 
                        "Give me a detailed explanation of the benchmark for "+alias+" for my portfolio.",
                        "How do I calculate the average for "+alias+" in my organization?",
                        "How do I calculate the average for "+alias+" in my org?",
                        "Go deeper about the benchmark for "+alias+" in my PMC",
                        "Can you go into more detail about the benchmark for "+alias+" in my property management company",
                        "Provide a thorough description and computation of the benchmark for "+alias]
        elif hierarchy == "Multifamily Level":
            question = ["Elaborate on the benchmark for "+alias+" in this industry", 
                        "Give me a detailed explanation of the benchmark for "+alias+" in multifamily.",
                        "How do I calculate the benchmark for "+alias+" in multifamily industry?",
                        "Provide a thorough description and computation of the multifamily benchmark for "+alias]
    elif metadata_column_name =="Godeeper.Example":
        if hierarchy == "Property Level":
            question = ["Show me a thorough demonstration of "+alias+" for my property", 
                        "Give me a detailed example of "+alias+" for my community",
                        "What is a clear and detailed illustration of "+alias+"?",
                        "Give me a comprehensive example of how "+alias+" is derived"]
        elif hierarchy == "PMC Level":
            question = ["Show me a thorough demonstration of the average of "+alias+" for my community", 
                        "Give me a detailed example of the benchmark for "+alias+" for my portfolio", 
                        "What is a clear and detailed illustration of the average of "+alias+" for my organization?",
                        "What is a clear and detailed illustration of the average of "+alias+" for my org?",
                        "Give me a comprehensive example of how the benchmark for "+alias+" is derived.",
                        "Can you give me an in-depth sample calculation of the benchmark for "+alias+" in my PMC?",
                        "Show an extensive illustration of the benchmark for "+alias+" in my property management company?"]
        elif hierarchy == "Multifamily Level":
            question = ["Show me a thorough demonstration of the benchmark for "+alias+" in multifamily", 
                        "Give me a detailed example of the benchmark for "+alias+" in this industry", 
                        "What is a clear and detailed illustration of the multifamily benchmark of "+alias+"?",
                        "Give me a comprehensive example of how the benchmark for "+alias+" is derived in the multifamily industry"]
    return question
```

```python
def get_metricCategorySelected(metric, hierarchy):
    if metric == 'Applied to Leased Duration':
        if hierarchy == 'Property Level':
            return "property-applied-to-leased-duration"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-applied-to-leased-duration"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-applied-to-leased-duration"
    elif metric == 'Applied to Leased Ratio':
        if hierarchy == 'Property Level':
            return "property-applied-to-leased-ratio"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-applied-to-leased-ratio"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-applied-to-leased-ratio"
    elif metric == 'Lead to Toured Duration':
        if hierarchy == 'Property Level':
            return "property-lead-to-toured-duration"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-lead-to-toured-duration"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-lead-to-toured-duration"
    elif metric == 'Lead to Toured Ratio':
        if hierarchy == 'Property Level':
            return "property-lead-to-toured-ratio"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-lead-to-toured-ratio"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-lead-to-toured-ratio"
    elif metric == 'Toured to Applied Duration':
        if hierarchy == 'Property Level':
            return "property-toured-to-applied-duration"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-toured-to-applied-duration"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-toured-to-applied-duration"
    elif metric == 'Toured to Applied Ratio':
        if hierarchy == 'Property Level':
            return "property-toured-to-applied-ratio"
        elif hierarchy == 'PMC Level':
            return "pmc-benchmark-for-toured-to-applied-ratio"
        elif hierarchy == 'Multifamily Level':
            return "multifamily-benchmark-for-toured-to-applied-ratio"
```
